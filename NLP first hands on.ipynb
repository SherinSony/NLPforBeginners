# Unsupervised clustering K-means

```python
from sklearn.cluster import KMeans
```

```python
#we will lemmatize each post in the original groups rather than using the individual words we got
cleaned1=[]
for pst in grp.data:
    cleaned1.append(' '.join([lem.lemmatize(wrd.lower()) for wrd in pst.split() if wrd.isalpha() and wrd not in allnames]))
```

```python
cleaned1[0]
```

```python
grp.data[0] #original
```

```python
k=KMeans(n_clusters=10)
```

```python
k.fit(c.fit_transform(cleaned1))
```

```python
lab=grp.target
```

```python
lab[lab > 2]
```

```python
plt.scatter(k.labels_,lab)
plt.ylabel('News group label #')
plt.xlabel('KMeans Clustering #')
plt.yticks(list(lab))
plt.xticks(k.labels_)
plt.show()  #we see many newsgroups repeated in many Kmeans Clusters ....Hmmmm...
```

```python
k.score(c.fit_transform(cleaned1))
```
